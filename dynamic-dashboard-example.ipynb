# Dynamic Dashboard Example  

**Dynamic dashboards** are dashboards that may change to adapt to changes in the data (such as listing available column selection) or react to the user's interaction (if user selects a specific option, further buttons with more specific information might appear).  

The following notebook contains an example of a use case of such **dynamic dashboards** for exploring a dataset or adding interactivity on reporting. For a simpler use case using non-dynamic dashboards refer to the [static dashboard example](https://mradmrglabedc01.svr.us.jpmchase.net:19231/user/r559042/notebooks/personal/r559042/dashboards/static-dashboard-example.ipynb).  

For a cleaner experience the IPy notebook only contains the minimal code necessary to render the dashboard.  

The dashboard is constructed in two separate pieces (small projects can be easily done within one, but it best to separate functionality from the actual point-and-click pieces):

* Dashboard functionality in [`functionality.py`](https://mradmrglabedc01.svr.us.jpmchase.net:19231/user/r559042/edit/personal/r559042/dashboards/dynamicDashboard/functionality.py): Data functionality and plotters   
* Dashboard creation [`gui.py`](https://mradmrglabedc01.svr.us.jpmchase.net:19231/user/r559042/edit/personal/r559042/dashboards/dynamicDashboard/gui.py): Creation of the actual dashboard
